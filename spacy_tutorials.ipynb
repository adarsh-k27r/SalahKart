```python
import spacy
nlp = spacy.load('en_core_web_sm')

text = "It's well said that, You never owe a cat but the cat does owe her master. And This is big sentence. "

doc = nlp(text)

print(doc)
```

```python
# Function to split text into sentences using spaCy
def split_into_sentences_spacy(text):
    return [sent.text for sent in doc.sents]
```

```python
op = split_into_sentences_spacy(text)
print(op)
```

```python
#Tokenization

tokens = [token.text for token in doc]
print(tokens)
```

```python
#Lemmatization

lemma = [token.lemma_ for token in doc]
print(lemma)
```

```python
# Part of Speech Tagging (POS)

poses = [token.pos_ for token in doc]
print(poses)
```

```python
# Named Entity Recognition (NER)

doc = nlp("Adarsh Kumar studies in Rajiv Gandhi Petroleum College from 2021 to 2025 in Computer Science and specializes in Python.")
for token in doc.ents:
  print(token.text, token.label_)
```
