```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline
import pandas as pd

# Load pre-trained model and tokenizer
model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForTokenClassification.from_pretrained(model_name)

# Create NER pipeline
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
# Define a function to extract skills
def extract_skills(text):
    # Use the NER pipeline to extract entities
    ner_results = ner_pipeline(text)
    print(ner_results)

    # Filter entities to get only the 'MISC' label which might include skills
    skills = [result['word'] for result in ner_results if result['entity'] in ['I-MISC', 'B-MISC']]

    return skills
```

```python

# Provide sample inputs.

resume_text = "Hii, I'm proficient in Java, Machine-learning, machine translation and jQuery"
job_desc_text = "Looking for machine learning specialist, java, and machine-translator with jQuery."


# Extract skills from texts
resume_skills = extract_skills(resume_text)
job_desc_skills = extract_skills(job_desc_text)

print("Extracted Resume Skills:", resume_skills)
print("Extracted Job Description Skills:", job_desc_skills)
```

```python
# Define a function to calculate matching percentage
def calculate_matching_percentage(resume_skills, job_desc_skills):
    match_count = len(set(resume_skills) & set(job_desc_skills))
    total_skills = len(set(job_desc_skills))
    if total_skills == 0:
        return 0
    return (match_count / total_skills) * 100
```

```python
# Calculate matching percentage
matching_percentage = calculate_matching_percentage(resume_skills, job_desc_skills)
print("Matching Percentage:", matching_percentage)
```
